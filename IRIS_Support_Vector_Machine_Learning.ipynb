```python
import pandas as pd
```

```python
data = pd.read_csv('/content/Iris.csv')
data.head()
```

```python
data.isnull()
```

```python
data.rename(columns = {'5.1':'Sepal length', '3.5':'Sepal width',
                              '1.4':'Petal length', '0.2':'Petal width', 'Iris-setosa':'Class'}, inplace = True)
data
```

```python
feature_col_names=['Sepal length', 'Sepal width', 'Petal length', 'Petal width']
predicted_class_name=['Class']
```

```python
data.keys()
```

```python
X=data[feature_col_names].values
Y=data[predicted_class_name].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test =train_test_split(X, Y, test_size=0.80, random_state=42, stratify=Y)

X_train
```

```python
from sklearn import svm
clf=svm.SVC(kernel='linear')
clf.fit(X_train,Y_train)
y_pred=clf.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:", metrics.accuracy_score(Y_test, y_pred))
```
